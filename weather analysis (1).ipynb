```python
string_cols = weather_df.select_dtypes(include=['object'])  # Select object type (strings)
string_cols = string_cols.apply(lambda x: x.str.replace("'", ""))  # Apply replace using str accessor
weather_df.update(string_cols)  # Update original DataFrame

```

```python
weather_df = weather_df[weather_df['Date'] != '2022-02-29']
```

```python
weather_df
```

```python
# change the data type of Date from object to datetime
weather_df["Date"] = pd.to_datetime(weather_df["Date"],format='mixed', dayfirst=True,errors='coerce')
```

```python
# remove the irrelevent data 
weather_df = weather_df[weather_df[' diff_pressure'] != '0.11700000000000088;']
```
